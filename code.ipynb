# QUESTION 2: 
#### APPLYING BASIC MACHINE LEARNING MODELS TO SOLVE PROBLEMS, INCLUDING MODELS WITHIN ENSEMBLE LEARNING.

```python
# Import module for machine learning models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Model 1: Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)
lr_pred = lr_model.predict(X_test)
print('Linear Regression - Mean Squared Error:', mean_squared_error(Y_test, lr_pred))

# Scatter plot
plt.scatter(Y_test, lr_pred, alpha=0.5)
plt.title('Linear Regression: Actual vs. Predicted Charges')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.show()
```

```python
from sklearn.tree import plot_tree
# Model 2: Decision Tree
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, Y_train)
dt_pred = dt_model.predict(X_test)
print('Decision Tree - Mean Squared Error:', mean_squared_error(Y_test, dt_pred))

# Create a Decision Tree with a limited depth for visualization
dt_model_vis = DecisionTreeRegressor(max_depth=3)
dt_model_vis.fit(X_train, Y_train)

# Plot the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(dt_model_vis, filled=True, feature_names=X.columns, rounded=True, class_names=['charges'])
plt.title('Decision Tree Visualization')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Model 3: AdaBoost
adb_model = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(), n_estimators=50, learning_rate=1.0, random_state=42)
adb_model.fit(X_train, Y_train)
adb_pred = adb_model.predict(X_test)
print('AdaBoost - Mean Squared Error:', mean_squared_error(Y_test, adb_pred))

# Extract feature importances from the AdaBoost model
feature_importances = adb_model.feature_importances_

# Get feature names
feature_names = X.columns

# Sort indices based on feature importances
indices = np.argsort(feature_importances)

# Create bar chart
plt.figure(figsize=(10, 6))
plt.barh(range(len(indices)), feature_importances[indices], align='center')
plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
plt.xlabel('Feature Importance')
plt.title('AdaBoost: Feature Importance')
plt.show()
```

```python
# Model 4: Random Forest (Ensemble Learning)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)
rf_pred = rf_model.predict(X_test)
print('Random Forest - Mean Squared Error:', mean_squared_error(Y_test, rf_pred))

# Extract feature importances from the Random Forest model
feature_importances_rf = rf_model.feature_importances_

# Sort indices based on feature importances
indices_rf = np.argsort(feature_importances_rf)

# Create bar chart for Random Forest
plt.figure(figsize=(10, 6))
plt.barh(range(len(indices_rf)), feature_importances_rf[indices_rf], align='center', color='skyblue')
plt.yticks(range(len(indices_rf)), [feature_names[i] for i in indices_rf])
plt.xlabel('Feature Importance')
plt.title('Random Forest: Feature Importance')
plt.show()
```

```python
# Model 5: Gradient Boosting
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train, Y_train)
gb_pred = gb_model.predict(X_test)
print('Gradient Boosting - Mean Squared Error:', mean_squared_error(Y_test, gb_pred))

# Extract feature importances from the Gradient Boosting model
feature_importances_gb = gb_model.feature_importances_

# Sort indices based on feature importances
indices_gb = np.argsort(feature_importances_gb) 
# Create a sorted bar chart for Gradient Boosting
plt.figure(figsize=(12, 8))
plt.barh(range(len(indices_gb)), sorted(feature_importances_gb), align='center', color='lightgreen')
plt.yticks(range(len(indices_gb)), [feature_names[i] for i in np.argsort(feature_importances_gb)])
plt.xlabel('Feature Importance')
plt.title('Gradient Boosting: Feature Importance (Sorted)')
plt.show()
```

```python
# Model 6: XGBoost
xgb_model = XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X_train, Y_train)
xgb_pred = xgb_model.predict(X_test)
print('XGBoost - Mean Squared Error:', mean_squared_error(Y_test, xgb_pred))

from xgboost import plot_importance

# Plot feature importance for XGBoost
plt.figure(figsize=(12, 8))
plot_importance(xgb_model, importance_type='weight', color='skyblue')
plt.title('XGBoost: Feature Importance')
plt.show()
```

```python
# Model 7: LightGBM
lgbm_model = LGBMRegressor(n_estimators=100, random_state=42)
lgbm_model.fit(X_train, Y_train)
lgbm_pred = lgbm_model.predict(X_test)
print('LightGBM - Mean Squared Error:', mean_squared_error(Y_test, lgbm_pred))

from lightgbm import plot_importance

# Plot feature importance for LightGBM
plt.figure(figsize=(12, 8))
plot_importance(lgbm_model, importance_type='split', color='lightcoral')
plt.title('LightGBM: Feature Importance')
plt.show()
```

# QUESTION 3: 
#### UTILIZE FEED FORWARD NEURAL NETWORK AND RECURRENT NEURAL NETWORK (OR A MODEL OF THIS TYPE) TO ADDRESS THE PROBLEM

```python
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Feed Forward Neural Network (FFNN)
ffnn_model = Sequential()
ffnn_model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
ffnn_model.add(Dense(32, activation='relu'))
ffnn_model.add(Dense(1, activation='linear'))
ffnn_model.compile(loss='mean_squared_error', optimizer='adam')

ffnn_model.fit(X_train_scaled, Y_train, epochs=10, batch_size=32, validation_split=0.2)

ffnn_pred = ffnn_model.predict(X_test_scaled)
print('Feed Forward Neural Network - Mean Squared Error:', mean_squared_error(Y_test, ffnn_pred))
```

```python
# Assuming you have already defined X_train_scaled and X_test_scaled as the scaled versions of your data
X_train_rnn = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_rnn = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Recurrent Neural Network (RNN)
rnn_model = Sequential()
rnn_model.add(LSTM(64, input_shape=(X_train_rnn.shape[1], X_train_rnn.shape[2])))
rnn_model.add(Dense(1, activation='linear'))
rnn_model.compile(loss='mean_squared_error', optimizer='adam')

rnn_model.fit(X_train_rnn, Y_train, epochs=10, batch_size=32, validation_split=0.2)

rnn_pred = rnn_model.predict(X_test_rnn)
print('Recurrent Neural Network - Mean Squared Error:', mean_squared_error(Y_test, rnn_pred))
```

# QUESTION 4: 
#### APPLY TECHNIQUES TO PREVENT OVERFITTING ON THE MODELS FROM QUESTION (2) AND QUESTION (3) TO ADDRESS THE PROBLEM

1. REGULARIZATION

```python
from keras.regularizers import l2

# Feed Forward Neural Network (FFNN)
ffnn_model = Sequential()
# Regularization được áp dụng cho trọng số của layer Dense đầu tiên trong FFNN.
ffnn_model.add(Dense(64, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=l2(0.01)))
# Regularization được áp dụng cho trọng số của layer Dense thứ hai trong FFNN.
ffnn_model.add(Dense(32, activation='relu', kernel_regularizer=l2(0.01)))
# Regularization được áp dụng cho trọng số của layer Dense cuối cùng (output layer) trong FFNN.
ffnn_model.add(Dense(1, activation='linear', kernel_regularizer=l2(0.01)))
ffnn_model.compile(loss='mean_squared_error', optimizer='adam')

ffnn_model.fit(X_train_scaled, Y_train, epochs=10, batch_size=32, validation_split=0.2)

ffnn_pred = ffnn_model.predict(X_test_scaled)
print('Feed Forward Neural Network - Mean Squared Error:', mean_squared_error(Y_test, ffnn_pred))

# Recurrent Neural Network (RNN)
rnn_model = Sequential()
rnn_model.add(LSTM(64, input_shape=(1, X_train_scaled.shape[1]), recurrent_regularizer=l2(0.01)))  # Regularization for LSTM
rnn_model.add(Dense(1, activation='linear', kernel_regularizer=l2(0.01)))  # Regularization for Dense layer
rnn_model.compile(loss='mean_squared_error', optimizer='adam')

rnn_model.fit(X_train_rnn, Y_train, epochs=10, batch_size=32, validation_split=0.2)

rnn_pred = rnn_model.predict(X_test_rnn)
print('Recurrent Neural Network - Mean Squared Error:', mean_squared_error(Y_test, rnn_pred))
```

2. DROPOUT

```python
from keras.layers import Dropout
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Feed Forward Neural Network (FFNN)
ffnn_model = Sequential()
# Dropout for FFNN
ffnn_model.add(Dropout(0.5))
ffnn_model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
ffnn_model.add(Dense(32, activation='relu'))
ffnn_model.add(Dense(1, activation='linear'))
ffnn_model.compile(loss='mean_squared_error', optimizer='adam')

ffnn_model.fit(X_train_scaled, Y_train, epochs=10, batch_size=32, validation_split=0.2)

ffnn_pred = ffnn_model.predict(X_test_scaled)
print('Feed Forward Neural Network - Mean Squared Error:', mean_squared_error(Y_test, ffnn_pred))

# Recurrent Neural Network (RNN)
rnn_model = Sequential()
# Dropout for RNN
rnn_model.add(Dropout(0.5))
rnn_model.add(LSTM(64, input_shape=(X_train_rnn.shape[1], X_train_rnn.shape[2])))  # Fix input_shape here
rnn_model.add(Dense(1, activation='linear'))
rnn_model.compile(loss='mean_squared_error', optimizer='adam')

rnn_model.fit(X_train_rnn, Y_train, epochs=10, batch_size=32, validation_split=0.2)

rnn_pred = rnn_model.predict(X_test_rnn)
print('Recurrent Neural Network - Mean Squared Error:', mean_squared_error(Y_test, rnn_pred))
```

3. EARLY STOPPING

```python
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Feed Forward Neural Network (FFNN)
ffnn_model = Sequential()
ffnn_model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
ffnn_model.add(Dense(32, activation='relu'))
ffnn_model.add(Dense(1, activation='linear'))
ffnn_model.compile(loss='mean_squared_error', optimizer='adam')
# Early stopping for FFNN
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
ffnn_model.fit(X_train_scaled, Y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

ffnn_pred = ffnn_model.predict(X_test_scaled)
print('Feed Forward Neural Network - Mean Squared Error:', mean_squared_error(Y_test, ffnn_pred))

# Recurrent Neural Network (RNN)
rnn_model = Sequential()
rnn_model.add(LSTM(64, input_shape=(X_train_rnn.shape[1], X_train_rnn.shape[2])))  # Fix input_shape here
rnn_model.add(Dense(1, activation='linear'))
rnn_model.compile(loss='mean_squared_error', optimizer='adam')
# Early stopping for RNN
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
rnn_model.fit(X_train_rnn, Y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

rnn_pred = rnn_model.predict(X_test_rnn)
print('Recurrent Neural Network - Mean Squared Error:', mean_squared_error(Y_test, rnn_pred))
```
